```python
import yfinance as yf
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
def ensure_directory(directory_path):
    if not os.path.exists(directory_path):
        os.makedirs(directory_path)

def download_data(tickers, period="3y", interval="1d"):
    stock_data = yf.download(tickers, period=period, interval=interval, group_by='ticker')
    return stock_data

def create_signals(data):
    all_data = []
    
    for ticker in data.columns.levels[0]:  
        ticker_data = data[ticker].copy()
        ticker_data['Ticker'] = ticker
        
        ticker_data['MA10'] = ticker_data['Close'].rolling(window=10).mean()
        ticker_data['MA50'] = ticker_data['Close'].rolling(window=50).mean()
        
        delta = ticker_data['Close'].diff(1)
        gain = delta.where(delta > 0, 0)
        loss = -delta.where(delta < 0, 0)
        avg_gain = gain.rolling(window=14).mean()
        avg_loss = loss.rolling(window=14).mean()
        rs = avg_gain / avg_loss
        ticker_data['RSI'] = 100 - (100 / (1 + rs))
        
        ticker_data['Price_Change'] = ticker_data['Close'].diff()  
        ticker_data['Target'] = (ticker_data['Price_Change'] > 0).astype(int) 
        
        ticker_data.dropna(inplace=True)
        all_data.append(ticker_data)
    
    return pd.concat(all_data) 

def eda(data, output_dir):
    ensure_directory(output_dir)

    with open(os.path.join(output_dir, "data_overview.txt"), "w") as f:
        f.write("Data Overview:\n")
        data.info(buf=f)
        f.write("\nSummary Statistics:\n")
        f.write(str(data.describe()))

    plt.figure(figsize=(10, 6))
    for ticker in data['Ticker'].unique():
        ticker_data = data[data['Ticker'] == ticker]
        plt.plot(ticker_data.index, ticker_data['Close'], label=ticker)
    plt.title('Stock Closing Prices Over Time')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.savefig(os.path.join(output_dir, "closing_prices.png"))
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.heatmap(data[['Open', 'High', 'Low', 'Close', 'Volume', 'MA10', 'MA50', 'RSI']].corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.savefig(os.path.join(output_dir, "correlation_heatmap.png"))
    plt.show()


def compute_performance_metrics(trades_df, output_dir):
    """
    Compute Sharpe ratio, cumulative returns, and other metrics from trade data.
    """
    trades_df['Daily Return'] = trades_df['Profit'] / 10000  
    
    trades_df['Cumulative Return'] = trades_df['Daily Return'].cumsum()
    
    sharpe_ratio = (trades_df['Daily Return'].mean() / trades_df['Daily Return'].std()) * np.sqrt(252)
    
    with open(os.path.join(output_dir, "performance_metrics.txt"), "w") as f:
        f.write(f"Sharpe Ratio: {sharpe_ratio:.2f}\n")
    
    return trades_df, sharpe_ratio


def export_pnl_chart(trades_df, output_dir):
    ensure_directory(output_dir)  

    plt.figure(figsize=(10, 6))
    plt.plot(trades_df['Date'], trades_df['Cumulative PnL'], label='Cumulative PnL', color='blue')
    plt.title('Cumulative Profit and Loss Over Time')
    plt.xlabel('Date')
    plt.ylabel('Cumulative PnL ($)')
    plt.legend()
    plt.grid(True)
    plt.savefig(os.path.join(output_dir, 'pnl_chart.png'))
    plt.show()

    plt.figure(figsize=(10, 6))
    plt.plot(trades_df['Date'], trades_df['Cumulative Return'], label='Cumulative Return', color='green')
    plt.title('Cumulative Return Over Time')
    plt.xlabel('Date')
    plt.ylabel('Cumulative Return')
    plt.legend()
    plt.grid(True)
    plt.savefig(os.path.join(output_dir, 'cumulative_return.png'))
    plt.show()

def backtest_and_evaluate(data, y_test, y_pred, output_dir):
    ensure_directory(output_dir) 

    accuracy = accuracy_score(y_test, y_pred)
    
    with open(os.path.join(output_dir, "accuracy_report.txt"), "w") as f:
        f.write(f'Accuracy: {accuracy:.2f}\n')
    
    with open(os.path.join(output_dir, "classification_report.txt"), "w") as f:
        f.write("\nClassification Report:\n")
        f.write(classification_report(y_test, y_pred))
    
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Down', 'Up'], yticklabels=['Down', 'Up'])
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.savefig(os.path.join(output_dir, "confusion_matrix.png"))
    plt.show()

    initial_cash = 10000
    total_profit = 0
    trades = []

    test_data = data.iloc[-len(y_test):]  
    
    for i in range(len(y_test)):
        trade_info = {}  
        trade_info['Date'] = test_data.index[i]
        trade_info['Open Price'] = test_data.iloc[i]['Open']
        trade_info['Close Price'] = test_data.iloc[i]['Close']
        
        if y_pred[i] == 1:  
            open_price = test_data.iloc[i]['Open']
            close_price = test_data.iloc[i]['Close']
            profit = close_price - open_price  
            trade_info['Action'] = 'Buy'
            trade_info['Profit'] = profit
            total_profit += profit
        else:
            trade_info['Action'] = 'Hold'
            trade_info['Profit'] = 0
        
        trade_info['Cumulative PnL'] = total_profit
        trades.append(trade_info)
    
    trades_df = pd.DataFrame(trades)
    
    trades_df, sharpe_ratio = compute_performance_metrics(trades_df, output_dir)

    export_pnl_chart(trades_df, output_dir)

    return trades_df, sharpe_ratio

def train_decision_tree(data, output_dir):
    X = data[['Open', 'High', 'Low', 'Close', 'Volume', 'MA10', 'MA50', 'RSI']]  
    y = data['Target']
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False, random_state=42)
    
    model = DecisionTreeClassifier(random_state=42)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    trades_df, sharpe_ratio = backtest_and_evaluate(data, y_test, y_pred, output_dir)
    
    trades_df.to_csv(os.path.join(output_dir, "trading_decisions.csv"), index=False)

    return model
```

```python
output_directory = "/Users/marshallinosunedi/Desktop/trading_results"

tickers = ['BTC-USD', 'ETH-USD']
stock_data = download_data(tickers, period='3y', interval='1d')
stock_data = create_signals(stock_data)

eda(stock_data, output_directory)

train_decision_tree(stock_data, output_directory)
```
